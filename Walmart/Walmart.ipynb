### Analytical & Insight-Driven Questions

#### 11. Which product category has the highest average profit margin? (Requires grouping by category and averaging profit_margin)

```python
Highest_Avergae =round(walmart.groupby('category')['profit_margin'].mean().sort_values(ascending=False).head(1)*100,2)

Highest_Avergae.astype('float').astype('str')+' %'
```

#### 12. Is there a correlation between the payment method (e.g., Cash, Ewallet, Credit card) and the customer rating? (Compare average rating grouped by payment_method)

```python
walmart.groupby('payment_method').agg(
    avg_rating=('rating', 'mean'),
    counts=('invoice_id', 'count'),
    max_value =('total','max'),
    min_value =('total','min')
).sort_values(['max_value','avg_rating'],ascending=False)
```

#### 13. Calculate the total profit for each transaction. Which transaction was the most profitable?

```python
walmart['total_profit']= walmart['total']*walmart['profit_margin']

#walmart.sort_values('total_profit',ascending=False).head(1)


walmart.loc[walmart['total profit'].idxmax()]
```

#### 14. What is the total sales volume (total) for each city? Which city is the top performer in this sample?

#### 15 On the available dates, which date had the highest number of transactions?



```python
walmart.groupby('date').agg(
    Transactions = ('invoice_id','count'),
    profit =('total profit','sum')
).sort_values(['Transactions','profit'], ascending= False).head()
```
